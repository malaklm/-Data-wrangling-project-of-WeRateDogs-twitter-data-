#### Tidiness
Issue1: df1 -> The dog stage stored in many column
Based on what I learned in this course the faster and effective way to handle the Pandas data frame problems is using Numpy package functions. From our data, the dog stage represent in 4 columns, each row has one stage and the others is none, So using a where() function is very useful here. <br />

Define Issue1: Create a column (DogeStage). Ectract not null value from each column. Drop (doggo,floofer,pupper,puppo)

```python
# Code
#Step1: Create a new column to store the values
df1_c['DogStage'] = None
```

```python
def fill_DogStage(df,col2,s_name):
    df['DogStage'] = np.where(col2 == s_name,col2,df1_c['DogStage'])
```

```python
#Step2: Replace data if doggo not None
fill_DogStage(df1_c,df1_c['doggo'],'doggo')
```

```python
#Step3: Replace data if floofer not None
fill_DogStage(df1_c,df1_c['floofer'],'floofer')
```

```python
#Step4: Replace data if pupper not None
fill_DogStage(df1_c,df1_c['pupper'],'pupper')
```

```python
#Step5: Replace data if puppo not None
fill_DogStage(df1_c,df1_c['puppo'],'puppo')
```

```python
# Test
df1_c.head(100)
```

```python
#Step6: Drop 4 column 
# Code
df1_c.drop(['doggo', 'floofer','pupper','puppo'], axis=1,inplace=True)

# Test 
df1_c.head(1)
```

Issue2: df1 -> There's two column for (rating) rating_numerator, rating_denominator. <br />

Define Issue2: df1 -> Create a column name (rating)equal to rating numerator/rating denominator. Drop rating numerator and rating denominator.

```python
# Code 
df1_c['rating'] = df1_c['rating_numerator']/df1_c['rating_denominator']

df1_c.drop(['rating_numerator', 'rating_denominator'], axis=1,inplace=True)

# Test
df1_c.head()
```

Issue3: df3 -> df2,df3 should be in df1. <be />
Define Issue3: Merge df1 with df2 and df3 (left join)

```python
# Code 
df3_c.rename(columns= {'id': 'tweet_id'},inplace=True)

merge_tables =  pd.merge(df3_c,df1_c, on='tweet_id', how='left')
merge_tables =  merge_tables.merge(df2_c, on='tweet_id', how='left')

# Test
merge_tables.head()
```
